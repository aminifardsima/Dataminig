```python
data['tipped'] = (data['tip_amount'] > 0).astype("int")
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data.columns
```

```python
data_p = data[['rate_code', 'passenger_count', 'trip_time_in_secs', 'trip_distance', 'fare_amount', 'surcharge', 'mta_tax', 'tolls_amount', 'total_amount']]
```

```python
data_p.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_p, data['tipped'])
```

```python
sc = StandardScaler()
sc = sc.fit(X_train)
data_scaled = sc.transform(X_train)
```

```python
data_scaled
```

```python
import pickle
with open('standard_scaler.kaviani', 'wb') as f:
    pickle.dump(sc, f)
```

```python
data_test_scaled = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
rf = rf.fit(data_scaled, y_train)
```

```python
rf.score(data_test_scaled, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(rf.predict(data_test_scaled), y_test))
```
